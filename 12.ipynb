# 맵시TI (Map-si TI)

> **설문 응답을 패션 취향 데이터로 해석하고, 실제 쇼핑 행동으로 연결하는 개인화 패션 추천 서비스**

맵시TI는 사용자의 설문 응답을 바탕으로 **퍼스널 컬러, 선호 스타일, 핏 성향, TPO 맥락**을 분석하고, 그 결과를 기반으로 **추천 페르소나·분석 요약·쇼핑 큐레이션**까지 한 번에 제공하는 웹 서비스입니다. 단순히 “인기 상품”을 보여주는 것이 아니라, 사용자의 답변을 **구조화된 프로필**로 변환하고, 데이터셋 기반 추천과 쇼핑몰 크롤링 결과를 결합해 **설명 가능한 추천 경험**을 만드는 데 초점을 두었습니다.

이 저장소는 포트폴리오 제출을 염두에 두고 정리된 버전으로,

- **프론트엔드**: 설문 입력, 결과 카드, 스타일 팁, 포인트 컬러, 쇼핑 큐레이션 UI
- **백엔드**: FastAPI API, 설문 파싱, 추천 로직, 응답 정규화
- **추천 엔진**: 스타일/컬러/핏 점수 계산, K-Means 기반 군집화 활용
- **데이터 수집**: 무신사·지그재그 크롤링, AI 유사도 재정렬

을 하나의 흐름으로 묶어 구현했습니다.

---

## 1. 프로젝트 목적

패션 추천 서비스는 종종 “많이 팔리는 상품”이나 “단순 카테고리 필터” 수준에서 멈추는 경우가 많습니다. 하지만 실제 사용자 경험은 훨씬 더 개인적입니다. 같은 상의를 찾더라도,

- 어떤 분위기를 선호하는지
- 어떤 핏이 편한지
- 어떤 색이 얼굴 톤과 잘 맞는지
- 어떤 상황(TPO)에서 입을 옷인지

에 따라 선택 기준이 달라집니다.

맵시TI는 이 문제를 **설문 기반 데이터 수집 → 사용자 취향 벡터화 → 추천 결과 해석 → 실제 상품 연결**의 구조로 풀고자 했습니다.

이 프로젝트의 목표는 다음과 같습니다.

1. **주관적인 스타일 취향을 데이터로 변환**한다.
2. **퍼스널 컬러와 스타일 성향을 함께 고려하는 추천 경험**을 만든다.
3. 결과를 단순 리스트가 아니라 **설명 가능한 UI**로 보여준다.
4. 추천 결과를 실제 상품 탐색과 연결해 **서비스형 프로토타입**으로 완성한다.
5. 프론트, 백엔드, 추천 로직, 크롤링까지 직접 연결하여 **엔드투엔드 프로젝트 경험**을 증명한다.

## 2. 프로젝트 개요

### 서비스 한 줄 소개
사용자의 설문 응답을 분석해 “나의 맵시TI”를 도출하고, 그에 맞는 패션 스타일 설명과 쇼핑 큐레이션을 제공하는 웹 애플리케이션입니다.

### 핵심 사용자 시나리오
1. 사용자가 설문 페이지에서 성별, 퍼스널 컬러, 스타일 문항에 응답합니다.
2. 프론트엔드가 설문 데이터를 백엔드 API(`/profile`, `/recommendations`)에 전달합니다.
3. 백엔드는 응답을 정규화해 사용자 프로필을 생성하고, 추천 알고리즘을 통해 상위 추천 결과를 계산합니다.
4. 결과 페이지에서 사용자는
   - 자신의 패션 페르소나,
   - 분석 요약,
   - 스타일 팁,
   - 포인트 컬러,
   - 카테고리별 쇼핑 큐레이션
   을 확인합니다.
5. 사용자가 **맞춤 상품 보기** 버튼을 눌렀을 때만 상품 추천을 로드하여, 첫 화면에 불필요한 더미 상품이 노출되지 않도록 UX를 설계했습니다.

- **Pytest**

설문 파서, 추천 서비스, 크롤러, API 엔드포인트, 통합 흐름 테스트가 포함되어 있습니다.

## 4. 핵심 기능

### 4-1. 설문 기반 사용자 프로필 생성
- 성별, 퍼스널 컬러, 스타일 응답을 입력받아 사용자 프로필을 생성합니다.
- `survey_parser.py`에서 응답을 정규화하고, 다음과 같은 정보를 계산합니다.
  - `personal_color`
  - `personal_color_display`
  - `primary_style`
  - `secondary_styles`
  - `fit_preference`
  - `style_search_keywords`
  - `color_search_keywords`

### 4-2. 스타일 성향 분류 및 추천 요약 생성
- 설문 응답을 스타일 점수 맵으로 변환합니다.
- K-Means 및 유사도 기반 추천 로직을 사용해 가장 유사한 스타일/시대감을 도출합니다.
- 결과 페이지에는 단순 문자열이 아니라 다음과 같은 **분석 요약 카드**로 제공합니다.
  - 분석 제목
  - 연대(예: 1990년대)
  - 스타일 라벨(예: 소피스티케이티드)
  - 유사도 퍼센트
  - 성별 / 퍼스널 컬러 / 핏 칩 UI

### 4-3. 설명 가능한 결과 화면 UX
결과 페이지는 추천 결과를 “상품 카드 나열”로 끝내지 않고, 사용자가 자신의 결과를 쉽게 이해할 수 있도록 다음 레이어로 나누어 구성했습니다.

- **나의 맵시TI**: 대표 페르소나와 캐릭터형 설명
- **분석 요약**: 알고리즘이 어떤 기준으로 사용자를 해석했는지 요약
- **스타일 팁**: 실제 코디에 바로 적용할 수 있는 문장형 팁
- **포인트 컬러**: 추천 컬러 칩
- **쇼핑 큐레이션**: 카테고리·핏·컬러 선택 후 맞춤 추천 확인

### 4-4. 버튼 클릭 기반 상품 노출
초기 결과 화면에서 더미 상품이 자동으로 노출되면 사용자에게 “이미 결과가 정해져 있었다”는 인상을 줄 수 있습니다. 이를 방지하기 위해,

- 결과 화면 진입 시에는 상품 추천 영역을 비워두고,
- 사용자가 **맞춤 상품 보기** 버튼을 눌렀을 때만
- 선택한 카테고리/핏/컬러 조합에 맞는 결과를 렌더링하도록 변경했습니다.

이 기능은 작은 UI 수정처럼 보이지만, **추천 서비스의 신뢰감**과 **인터랙션 흐름의 자연스러움** 측면에서 중요한 개선 포인트였습니다.

### 4-5. 크롤링 기반 실상품 추천
- 무신사, 지그재그 대상 크롤러가 준비되어 있습니다.
- 클래스명이 자주 바뀌는 동적 페이지 구조를 고려해, **이미지 → 부모 노드 역추적 → 텍스트 정제** 방식으로 상품 정보를 수집합니다.
- 크롤링 후보군을 AI 유사도 재평가로 정렬해 더 적합한 결과를 상위에 배치합니다.

## 5. 시스템 아키텍처

```text
[User]
  ↓
[Frontend: index.html / survey.html / result.html]
  ↓  POST /profile
  ↓  POST /recommendations
[FastAPI Backend]
  ├─ survey_parser.py          : 설문 응답 정규화
  ├─ fashion_config.py         : 스타일/컬러/핏 매핑 규칙
  ├─ recommender.py            : 유사도 계산 및 랭킹
  ├─ recommendation_service.py : 분석 요약 및 추천 응답 조립
  ├─ crawlers/                 : 무신사 / 지그재그 크롤링
  └─ ai_similarity_service.py  : AI 유사도 재정렬
  ↓
[Response]
  ├─ user_profile
  ├─ preference_analysis
  ├─ analysis_summary
  └─ recommendation_results
  ↓
[Frontend Result UI Rendering]
```

### 요청 흐름
1. `survey.html`에서 설문 완료
2. `frontend/js/api.js`에서 백엔드 요청용 payload 생성
3. `POST /profile`로 사용자 프로필 분석
4. `POST /recommendations`로 추천 결과 및 분석 요약 수신
5. 응답을 `sessionStorage`에 저장
6. `result.html`에서 `result.js`가 결과 화면 렌더링
7. 사용자가 버튼 클릭 시 쇼핑 큐레이션 영역 렌더링

## 7. 설치 및 실행 방법

### 7-1. 권장 실행 환경
- Python 3.10 이상
- pip
- 가상환경(venv) 권장
- 실시간 크롤링 기능 사용 시 Chrome 계열 브라우저 환경 필요 가능

### 7-2. 로컬 실행 (권장)
현재 저장소는 **FastAPI가 프론트엔드 정적 파일까지 직접 서빙하는 방식**을 기준으로 가장 안정적으로 동작합니다.

```bash
cd backend
python -m venv .venv
source .venv/bin/activate
pip install -r requirements.txt
uvicorn app.main:app --reload --host 0.0.0.0 --port 8000
```

브라우저에서 아래 주소로 접속합니다.

```bash
http://localhost:8000
```

### 7-3. 테스트 실행
```bash
cd backend
pytest tests -q
```

테스트는 다음 범위를 포함합니다.
- 설문 파서 단위 테스트
- 추천 서비스 테스트
- 크롤러 테스트
- API 엔드포인트 테스트
- 통합 테스트

### 7-4. Docker Compose에 대한 메모
`docker-compose.yml`이 포함되어 있지만, 현재 포트폴리오 제출용 정리 버전은 **단일 FastAPI 서버 구동 방식**을 기준으로 설명하는 것이 가장 정확합니다. 프론트/백엔드 분리 배포는 추가 보완 후 사용하는 것을 권장합니다.

## 10. 개발 과정에서 겪은 이슈와 해결 과정

이 프로젝트에서 가장 중요하게 생각한 부분은 단순히 “기능을 만든다”가 아니라, **서비스가 실제로 동작하는 흐름을 끝까지 연결하는 것**이었습니다. 구현 과정에서 마주친 핵심 이슈와 해결 과정을 정리하면 다음과 같습니다.

### 이슈 1. 프론트엔드와 백엔드 포트 불일치로 API 호출 실패
**문제**
- 프론트엔드와 백엔드가 서로 다른 포트에서 실행될 때, API 요청이 잘못된 주소로 향하거나 세션에 저장된 오래된 주소를 참조하는 문제가 발생했습니다.
- 그 결과 `/profile`, `/recommendations` 호출이 실패하고, 결과 화면이 비정상적으로 표시됐습니다.

**해결**
- `utils.getApiBaseUrl()`을 중심으로 API 주소를 일관되게 계산하도록 정리했습니다.
- 백엔드가 프론트엔드 정적 파일을 직접 서빙하도록 수정해, 최종적으로는 **단일 서버 구동 방식**으로 안정화했습니다.
- 세션 스토리지에 남아 있는 이전 포트 정보로 인해 오작동하는 상황도 예외 처리했습니다.

**성과**
- 개발 환경 차이로 인한 통신 오류가 크게 줄었고,
- 포트폴리오 제출용 데모를 한 번의 실행으로 재현할 수 있게 되었습니다.

### 이슈 2. 백엔드 스타일 코드와 프론트엔드 페르소나 라벨이 달라 결과 UI 매칭 실패
**문제**
- 백엔드는 `modern_minimal`, `romantic` 같은 내부 스타일 코드를 사용하고,
- 프론트엔드는 `모던/미니멀`, `로맨틱` 같은 한글 라벨을 기준으로 페르소나 객체를 찾고 있었습니다.
- 이 불일치 때문에 결과 카드가 잘못 표시되거나 기본값으로 fallback되는 문제가 있었습니다.

**해결**
- `STYLE_LABELS`를 기준으로 백엔드 응답 직전에 스타일 코드를 프론트가 이해할 수 있는 라벨로 변환했습니다.
- 동시에 `analysis_summary` 구조를 별도로 만들어, 결과 화면이 필요한 값을 명확한 키로 받을 수 있도록 응답 스키마를 정리했습니다.

**성과**
- 결과 카드와 페르소나 이미지, 스타일 팁이 안정적으로 일치하게 되었고,
- 프론트엔드가 “억지로 추론”하지 않아도 되도록 데이터 계약(contract)을 개선했습니다.

### 이슈 3. 결과 화면에 자동으로 보이는 더미 상품이 사용자 신뢰를 떨어뜨림
**문제**
- 결과 페이지에 진입하자마자 더미 상품이 먼저 보이면, 사용자는 “버튼을 누르지 않아도 결과가 이미 정해져 있었다”는 인상을 받을 수 있었습니다.
- 추천 서비스에서 이런 경험은 신뢰도 저하로 이어질 수 있습니다.

**해결**
- 초기 렌더링 시 `product-results`, `shop-links`를 비워두고,
- **맞춤 상품 보기** 버튼 클릭 시에만 상품을 렌더링하도록 흐름을 수정했습니다.

**성과**
- 추천 결과 노출 타이밍이 사용자 액션과 맞물리게 되었고,
- 결과 화면의 인터랙션 구조가 훨씬 자연스러워졌습니다.

### 이슈 4. 크롤링 대상 사이트의 동적 구조 변화로 셀렉터 의존 방식이 자주 깨짐
**문제**
- 무신사와 지그재그는 화면 구조나 클래스명이 바뀌는 경우가 많아, 전형적인 CSS selector 기반 크롤링이 쉽게 깨졌습니다.
- 지그재그는 광고, 배너, 추적용 이미지, 앱 유도 요소가 섞여 있어 상품만 정제해서 추출하기가 까다로웠습니다.

**해결**
- **이미지 노드를 먼저 찾은 뒤 부모 노드를 역추적해 상품 정보를 수집하는 방식**으로 변경했습니다.
- 광고/시스템 문구를 제거하는 블랙리스트와 정규식 정제 로직을 강화했습니다.
- 지연 로딩 대응을 위한 대기 및 스크롤 전략도 조정했습니다.

**성과**
- 사이트 구조 변경에 덜 취약한 수집 로직을 확보했고,
- 노이즈가 많은 결과에서 실제 상품만 추출하는 정밀도가 개선되었습니다.

### 이슈 5. 데이터셋 이미지 경로와 실제 정적 파일 경로가 맞지 않아 이미지 404 발생
**문제**
- CSV 기준 이미지 경로와 실제 저장 경로가 맞지 않는 경우가 있어, 추천 결과에 이미지가 있어도 화면에 보이지 않는 문제가 있었습니다.

**해결**
- `item_id` 또는 파일명을 기준으로 데이터셋 디렉터리를 재탐색(`rglob`)해 실제 파일 위치를 추적하도록 보정했습니다.
- 그 결과 정적 URL(`/sample-files/...`)을 동적으로 생성할 수 있도록 했습니다.

**성과**
- 추천 결과의 시각적 완성도가 높아졌고,
- 데모 품질과 사용성 모두 개선되었습니다.

### 이슈 6. 실시간 AI 유사도 평가를 넣자 응답 속도가 급격히 느려짐
**문제**
- 외부 AI 서버에 상품 이미지를 하나씩 순차적으로 보내면 응답 시간이 길어져, 실사용 데모에서 기다림이 크게 느껴졌습니다.

**해결**
- `ThreadPoolExecutor`를 사용해 다수의 후보 상품 이미지를 병렬로 평가하도록 바꿨습니다.
- 후보군을 먼저 20개 내외로 제한한 뒤 재정렬하는 방식으로, 속도와 품질 사이의 균형을 맞췄습니다.

**성과**
- 응답 시간을 실사용 가능한 수준으로 단축했고,
- 추천 품질을 유지하면서도 UX를 해치지 않는 구조를 만들 수 있었습니다.

## 11. 테스트와 검증 포인트

이 프로젝트는 UI 구현에만 집중하지 않고, 실제 로직이 망가지지 않도록 테스트 코드를 함께 유지했습니다.

특히 다음을 검증합니다.

- 설문 입력이 서로 다른 프로필을 올바르게 생성하는지
- 성별/스타일/핏 조건에 따라 필터링 및 랭킹이 올바르게 동작하는지
- 추천 검색 프로필이 무신사·지그재그 크롤링 맥락에 맞게 만들어지는지
- FastAPI 엔드포인트가 정상 응답을 반환하는지
- 내부 파이프라인 전체가 mock 데이터셋 기준으로 일관되게 실행되는지

포트폴리오 관점에서 이 부분은 “단순 구현”을 넘어, **서비스 완성도를 검증하는 개발 습관**을 보여주는 중요한 요소라고 생각합니다.

---

## 12. 프로젝트를 통해 얻은 인사이트

이 프로젝트를 통해 얻은 가장 큰 인사이트는 다음 두 가지였습니다.

### 1) 추천 서비스는 “정답”보다 “설명 가능성”이 중요하다
사용자가 납득할 수 없는 추천은 아무리 점수가 높아도 좋은 경험이 되기 어렵습니다. 그래서 맵시TI에서는 분석 요약, 스타일 팁, 포인트 컬러처럼 **결과를 해석할 수 있는 레이어**를 별도로 설계했습니다.

### 2) 실제 서비스 품질은 작은 UX 디테일에서 갈린다
결과 카드 위치, 폰트 크기, 배경색, 상품 노출 시점처럼 사소해 보이는 부분이 사용자 인식에는 큰 차이를 만듭니다. 이번 작업에서는 백엔드 응답 구조와 프론트 UI를 함께 조정하며, **데이터와 화면을 동시에 설계하는 경험**을 쌓을 수 있었습니다.

## 13. 향후 개선 방향

- 추천 결과에 대한 **사용자 피드백 수집(좋아요/별로예요)** 기능 추가
- 설문 기반 추천과 클릭 로그를 함께 사용하는 **하이브리드 추천 구조**로 확장
- 결과 화면의 공유 기능 고도화
- 카테고리별 추천 이유를 더 세밀하게 텍스트로 생성하는 **설명형 추천 시스템** 확장
- Docker 기반 배포 구조 및 CI 파이프라인 정리

---

## 14. 마무리

맵시TI는 단순한 UI 시안이 아니라, **설문 기반 추천 로직, 실제 상품 수집, 결과 해석 UI, 인터랙션 개선**까지 포함한 패션 추천 서비스 프로토타입입니다.

포트폴리오 제출 관점에서 이 프로젝트가 보여주는 강점은 다음과 같습니다.

- 문제 정의부터 UI/백엔드/데이터 흐름까지 직접 설계한 점
- 추천 시스템을 설명 가능한 형태로 구현한 점
- 크롤링, 응답 정규화, 테스트, UX 개선까지 실제 서비스 관점으로 다룬 점

즉, “한 화면을 예쁘게 만드는 것”을 넘어, **사용자 입력이 어떻게 데이터가 되고, 그 데이터가 어떻게 추천 경험으로 이어지는지**를 끝까지 구현한 프로젝트입니다.
